### This workbook downloads the stock data for given duration and interval.

### Import libraries

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from datetime import datetime
import plotly.express as px
```

```python
df = pd.read_csv('../data/all_stock.csv')
```

```python
def show_plot(df,start_date,end_date):
    df['date_only'] = df['Datetime'].apply(lambda x : x[:10])
    temp_df = df[(df['date_only'] >= start_date) & (df['date_only'] <= end_date)].reset_index(drop=True)

    fig = px.line(temp_df, x = 'Datetime', y = ['High','Low','Close','Open'])
    fig.show()
```

```python
start_date = '2024-08-08'
end_date = '2024-08-08'

show_plot(df,start_date,end_date)
```
